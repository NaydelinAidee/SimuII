```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Configuracin de parmetros
n_pasos = 100
W = np.zeros(n_pasos + 1)  # Creamos un vector de ceros para guardar los valores

# 2. Simulacin de la trayectoria
# El primer valor W[0] ya es 0. Empezamos a llenar desde el 1.
for t in range(1, n_pasos + 1):
    z = np.random.normal(0, 1) # Generamos el incremento aleatorio Z ~ N(0,1)
    W[t] = W[t-1] + z          # El nuevo valor es el anterior + el ruido

# 3. Graficar el resultado
plt.figure(figsize=(10, 5))
plt.plot(W, label="Trayectoria de Wt", color="blue", linewidth=1.5)
plt.axhline(0, color='black', lw=1, ls='--') # Lnea de referencia en cero
plt.title("Simulacin de un Proceso Estocstico (Movimiento Browniano)")
plt.xlabel("Tiempo (t)")
plt.ylabel("Valor (Wt)")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Configuracin de parmetros
n_pasos = 100
n_trayectorias = 5
tiempo = np.arange(n_pasos + 1)

# 2. Crear la figura
plt.figure(figsize=(12, 6))

# 3. Bucle para generar 5 trayectorias distintas
for i in range(n_trayectorias):
    # Generamos 100 incrementos aleatorios Z ~ N(0, 1)
    incrementos = np.random.normal(0, 1, n_pasos)

    # Creamos el vector W empezando en 0
    W = np.zeros(n_pasos + 1)
    # W_t es la suma acumulada de los incrementos (la integral del ruido)
    W[1:] = np.cumsum(incrementos)

    # Graficamos cada trayectoria con un color distinto (automtico)
    plt.plot(tiempo, W, label=f"Trayectoria {i+1}")

# 4. Detalles del grfico
plt.axhline(0, color='black', lw=1, ls='--') # Lnea de referencia en cero
plt.title("Inciso b): 5 Trayectorias del Proceso $W_t$")
plt.xlabel("Tiempo (t)")
plt.ylabel("Valor ($W_t$)")
plt.legend()
plt.grid(True, alpha=0.3)
plt.savefig("trayectorias_brownianas.png")
plt.show()
```
