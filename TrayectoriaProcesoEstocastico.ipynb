```python
from random import gauss
import matplotlib.pyplot as plt

def gen(m):
    l = [0]  # Lista inicial con el valor de origen
    w = 0    # Valor inicial de la caminata
    for i in range(m):
        z = gauss(0, 1)  # Generamos el choque aleatorio (ruido blanco)
        w = w + z        # Actualizamos el valor actual (acumulacin)
        l.append(w)      # Guardamos el nuevo valor en la lista
    return l

# Parmetros de la simulacin
m = 100  # Nmero de pasos en el tiempo

# Generar y graficar 5 trayectorias diferentes
for i in range(5):
    trayectoria = gen(m)
    plt.plot(trayectoria)

plt.show()
```

```python
from random import gauss
from statistics import mean, variance
from math import sqrt

def f(t, num_simulaciones):
    l = []
    for i in range(num_simulaciones):
        # W_t ~ N(0, t) se simula como sqrt(t) * N(0, 1)
        wt = sqrt(t) * gauss(0, 1)
        l.append(wt**2) # Guardamos el cuadrado
    return l

# Lista de tiempos a evaluar
T = [0.1, 0.01, 0.001, 0.0001, 0.00001]
M = 1000000 # Un milln de simulaciones

print(f"{'Tiempo (t)':<12} {'Media (esperada ~t)':<20} {'Varianza (esperada ~2t^2)':<20}")
print("-" * 60)

for t in T:
    lista_valores = f(t, M)

    media_calc = mean(lista_valores)
    varianza_calc = variance(lista_valores)

    # Formato similar al print de C (%f) que usaste en el pizarrn
    print(f"{t:<12.5f} {media_calc:<20.8f} {varianza_calc:<20.8f}")
```
